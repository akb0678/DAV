# Plotting

```python
# NO. of wiining seasons of a team
f_w = ipl[ipl['MatchNumber']=='Final']['WinningTeam'].value_counts()
teams =  ipl[ipl['MatchNumber']=='Final']['WinningTeam'].value_counts().index
plt.pie(f_w,labels = teams )



# Adding a legend to the pie chart
plt.legend(title='Teams', loc='best', bbox_to_anchor=(2, 0.5))

# Adding a title for the plot
plt.title('Number of Winning Seasons for Each Team')

plt.show()
```

```python
# Season wise performance of RCB
x1=ipl[ipl['WinningTeam']=='Royal Challengers Bangalore']['Season'].value_counts().sort_index()
lbx=x1.index

ipl['lx']=pd.Series(lbx)

lby=x1.values
ipl['ly']=pd.Series(lby)
plt.figure(figsize=(15, 6))
sns.lineplot(data=ipl,x='lx',y='ly')

plt.xlabel('Season')
plt.ylabel('Number of Wins')
plt.title('Number of Wins for Royal Challengers Bangalore over Seasons')

```
